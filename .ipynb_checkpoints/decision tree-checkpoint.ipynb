# 분석_ 머신러닝

```python
df_mc = df.drop(columns = ["adult_male", "embark_town","embarked","alive"])
df_mc.info()
```

```python
df.head(3)
```

```python
print(df_mc["sex"].unique())
print(df_mc["class"].unique())
print(df_mc["who"].unique())
print(df_mc["alone"].unique())
```

```python
sex_mapping = {"male" : 0, "female" : 1}
df_mc["sex"] = df_mc["sex"].map(sex_mapping) # 맵핑을 할 것이라는 함수

cls_mapping = {"First" : 1, "Second" : 2, "Third" : 3}
df_mc["class"] = df_mc["class"].map(cls_mapping)

who_mapping = {"man" : 1, "woman" : 2, "child" : 3}
df_mc["who"] = df_mc["who"].map(who_mapping)

aln_mapping = {False : 0, True : 1}
df_mc["alone"] = df_mc["alone"].map(aln_mapping)
```

```python
df_mc
```

```python
df_mc.columns
```

## train_test_split()
### test_size 테스트테스트 셋으로 저장할 확률
### random_state : seed 넘버 입력. 값을 할당하지 않으면 매번 새로운 데이터로 랜덤 추출하여 데이터 셋이 변경

```python
from sklearn.model_selection import train_test_split
seed = 1004
train, test = train_test_split(df_mc, test_size = 0.2, random_state = seed)
```

```python
train # 훈련 데이터 80%
```

```python
test # 테스트 데이터 20%
```

```python
train_x = train.drop(["survived"], axis = 1) # 독립변수 (원인)
train_y = train["survived"] # 종속 변수 (결과)
```

```python
test_x = test.drop(["survived"], axis = 1) # 독립변수 (원인)
test_y = test["survived"] # 종속 변수 (결과)
```

```python
train_x.info()
```

```python
train_y
```

# Decision Tree modeling

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(max_depth=3, random_state = 1004) # 가지 최대가지수 3개
tree_clf.fit(train_x, train_y) # 독립변수 , 종속변수
tree_clf.score(train_x, train_y)
```

## export_graphviz() : 의사결정나무를 시각화

```python
import os 
os.environ["PATH"]+=os.pathsep+'E:/Users/bo/anaconda3/Library/bin/graphviz'
```

```python
from sklearn.tree import export_graphviz
import graphviz
feature_names = test_x.columns # ["pclass", "sex","age","sibsp","parch","fare","embarked"] ...
export_graphviz(tree_clf,
               filled = True, # class 별 색상
               feature_names = feature_names,
               class_names = ["depth", "Survived"],
               out_file = "decision-tree.dot"
               )

with open("decision-tree.dot") as f: # 시각화를 위한 변수 할당
    dot_graph = f.read()
    
graphviz.Source(dot_graph)
```
